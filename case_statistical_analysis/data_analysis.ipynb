```python
import pandas as pd
#Importando o CSV
df_csv = pd.read_csv('C:/Área de Trabalho/UniCeub/Sistematizacao_mat.csv', sep = ';')
```

```python

df_csv.shape[0]
df_csv.info()
df_csv.drop_duplicates()
df_csv.head(100)

```

```python
# Group by com Count distinct nas variaveis categoricas
dfNome = pd.DataFrame(data = df_csv)
dfNome = dfNome.groupby('Nome')['#'].nunique()

dfGenero = pd.DataFrame(data = df_csv)
dfGenero = dfGenero.groupby('Gênero')['#'].nunique()

dfNacionalidade = pd.DataFrame(data = df_csv)
dfNacionalidade = dfNacionalidade.groupby('Nacionalidade')['#'].nunique()

dfLocaldeResidencia = pd.DataFrame(data = df_csv)
dfLocaldeResidencia = dfLocaldeResidencia.groupby('Local de residência')['#'].nunique()

dfProfissao = pd.DataFrame(data = df_csv)
dfProfissao = dfProfissao.groupby('Profissão')['#'].nunique()

dfEducacao = pd.DataFrame(data = df_csv)
dfEducacao = dfEducacao.groupby('Nível de educação')['#'].nunique()


dfEstadoCivil = pd.DataFrame(data = df_csv)
dfEstadoCivil = dfEstadoCivil.groupby('Estado civil')['#'].nunique()

dfAreaInteresse = pd.DataFrame(data = df_csv)
dfAreaInteresse = dfAreaInteresse.groupby('Área de interesse')['#'].nunique()

dfMeioTransporte = pd.DataFrame(data = df_csv)
dfMeioTransporte = dfMeioTransporte.groupby('Meio de transporte utilizado')['#'].nunique()

dfMidea = pd.DataFrame(data = df_csv)
dfMidea = dfMidea.groupby('Uso de mídias sociais')['#'].nunique()

print(dfGenero)
print(dfNacionalidade)
print(dfLocaldeResidencia)
print(dfProfissao)
print(dfEducacao)
print(dfEstadoCivil)
print(dfAreaInteresse)
print(dfMeioTransporte)
print(dfMidea)
```

```python
# Converte o % da coluna Número de filhos em inteiro
df_csv['Número de filhos'] = df_csv['Número de filhos'].str.replace('%', '').str.replace(',00', '').astype(int) / 100
df_csv
```

```python
# Count, Média, Minino, Máximo, Desvio Padrão
dfIdade = df_csv['Idade'].describe()
dfRenda = df_csv['Renda anual'].describe()
dfFilhos = df_csv['Número de filhos'].describe()
dfHoras = df_csv['Horas de trabalho por semana'].describe()
dfSatisfacao = df_csv['Nível de satisfação com o trabalho atual'].describe()

print(dfIdade)
print(dfRenda)
print(dfFilhos)
print(dfHoras)
print(dfSatisfacao)
```

```python
# Mediana das colunas
print(df_csv.median(numeric_only=True))

# Moda das colunas
print(df_csv.mode(numeric_only=True))
```

```python
import matplotlib.pyplot as plt
# Crie o gráfico de dispersão
df_csv.plot.scatter(x='Idade', y='Renda anual')
plt.xlabel('Idade')
plt.ylabel('Renda anual')
plt.title('Gráfico de Dispersão')
plt.show()

# Crie o gráfico de dispersão
df_csv.plot.scatter(x='Renda anual', y='Idade')
plt.xlabel('Renda anual')
plt.ylabel('Idade')
plt.title('Gráfico de Dispersão')
plt.show()
```
